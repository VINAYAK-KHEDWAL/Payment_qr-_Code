```python
import qrcode
```

```python
#  Taking upi id as input
upi_id =  input("Enter your UPI ID = ")
```

```python
# upi://pay?pa=UPI_ID&pn=NAME&am=Amount&cu=CURRENCY&tn=MESSAGES 
```

```python
# Defining thr  URL based on the UPI ID and the payment App
# You  can Modify these URLs bansed on thr payment apps you want to support 
```

```python
phonepe_url = f'upi://pay?pa={upi_id}&pn=Recipient%20Name&mc=1234' 
paytm_url = f'upi://pay?pa={upi_id}&pn=Recipient%20Name&mc=1234' 
google_pay_url = f'upi://pay?pa={upi_id}&pn=Recipient%20Name&mc=1234' 
```

```python
# create qr codes for each payment app
phonepe_url = qrcode.make(phonepe_url)
paytm_url = qrcode.make(paytm_url)
google_pay_url = qrcode.make(google_pay_url)
```

```python
# save the qr code to image file 
For Phone Pe
phonepe_url.save('Phonepe_qr.png')
```

```python
# save the qr code to image file 
For Paytm

paytm_url.save('paytm_qr.png')
```

```python
# save the qr code to image file 
For Google Pay

google_pay_url.save('google_pay_qr.png')
```
